# 2026 FIFA World Cup Predictor

## Project Goal
Predict group stage qualifiers and knockout stage match winners
for the 2026 FIFA World Cup.

## Two-Stage Prediction System
1. **Group Stage** — predict which 2 teams (+ best 3rd place) qualify per group
2. **Knockout Stage** — predict match winners round by round

## Model Architecture
- Match Score Model (Bivariate Poisson)
- Group Stage Monte Carlo Simulator
- Knockout Bracket Simulator
